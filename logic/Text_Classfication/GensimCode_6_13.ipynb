```python
#사용할 gensim 모델 생성

model = gensim.models.doc2vec.Doc2Vec(vector_size = 100 , min_count = 20, window = 5, epochs = 100, dm = 1, workers = 1, seed = 1)
```

```python
model.build_vocab(tag_data)
```

```python
print(model.iter)
```

```python
model.train(tag_data, total_examples = model.corpus_count, epochs = model.iter)
```

```python
model.alpha -=0.002
model.min_alpha = model.alpha
model.save("d2v.model")
print("Model Saved")
```
